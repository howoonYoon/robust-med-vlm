# Utility

```python
import os
import pandas as pd
import numpy as np

# ============================
# 0. Í≤ΩÎ°ú ÏÑ§Ï†ï
# ============================
BASE_OUT_DIR = os.path.expanduser("~/Lectures/Research_Project/Codes/PromptEngineering/Results")
METRICS_CSV  = os.path.join(BASE_OUT_DIR, "all_metrics.csv")

print(f"[INFO] Loading metrics from: {METRICS_CSV}")
df = pd.read_csv(METRICS_CSV)


# Í∏∞Î≥∏ Ï†ïÍ∑úÌôî
df["severity"] = df["severity"].astype(str).str.lower()
df["modality"] = df["modality"].astype(str).str.lower()
df["run_name"] = df["run_name"].astype(str)
df["model"]    = df["model"].astype(str)

print("‚úÖ unique run_name:", df["run_name"].unique())
print("‚úÖ unique model:", df["model"].unique())
```

```python
import matplotlib.pyplot as plt
import numpy as np

prompt_order = [
    "clean_image",  # Ïö∞Î¶¨Í∞Ä Ï∂îÍ∞ÄÌïú Ìñâ
    "baseline-prompts",
    "disease-cue-1-prompts",
    "disease-cue-reason-1-prompts",
    "artefact-disease-cue-1-prompts",
    "artefact-disease-cue-2-prompts",
    "artefact-disease-cue-reason-1-prompts",
]


def plot_heatmap(df, title):
    """
    df: index = run_name, columns = modality or model
    """
    data = df.values
    row_labels = df.index.tolist()
    col_labels = df.columns.tolist()

    fig, ax = plt.subplots(figsize=(1.2 * len(col_labels), 0.6 * len(row_labels) + 2))

    im = ax.imshow(data, aspect='auto')  # ÏÉâÏÉÅÏùÄ Í∏∞Î≥∏Í∞í ÏÇ¨Ïö©

    # Ï∂ï ÎùºÎ≤®
    ax.set_xticks(np.arange(len(col_labels)))
    ax.set_xticklabels(col_labels, rotation=45, ha="right")
    ax.set_yticks(np.arange(len(row_labels)))
    ax.set_yticklabels(row_labels)

    ax.set_title(title)
    fig.colorbar(im, ax=ax)

    plt.tight_layout()
    plt.show()
    
```

```python
import pandas as pd
import os

BASELINE_CSV = os.path.join(BASE_OUT_DIR, "baseline_performance.csv")

baseline = pd.read_csv(BASELINE_CSV)

# Ï†ïÎ¶¨
baseline["model"] = baseline["model"].astype(str)
baseline["modality"] = baseline["modality"].astype(str).str.lower()
baseline["metric"] = baseline["metric"].astype(str).str.lower()

display(baseline.head())
```

# Weak artefact performance

### Setting

```python
clean_acc = (
    baseline[baseline["metric"]=="accuracy"]
    .pivot_table(
        index="run_name",      # Í∑∏ÎÉ• Ìïú Ï§ÑÏùº Í∞ÄÎä•ÏÑ± ÎÜíÏùå
        columns="modality",
        values="value"
    )
)

# modalityÎ≥Ñ mean (run_name Ï†úÍ±∞)
clean_acc_row = clean_acc.mean(axis=0)

print("üßº Clean accuracy row")
display(clean_acc_row)

clean_f1 = (
    baseline[baseline["metric"]=="f1"]
    .pivot_table(
        index="run_name",
        columns="modality",
        values="value"
    )
)

clean_f1_row = clean_f1.mean(axis=0)

print("üßº Clean F1 row")
display(clean_f1_row)

clean_recall = (
    baseline[baseline["metric"]=="recall"]
    .pivot_table(
        index="run_name",
        columns="modality",
        values="value"
    )
)

clean_recall_row = clean_recall.mean(axis=0)

print("üßº Clean Recall row")
display(clean_recall_row)


# ---------------------Model ------------------------
clean_acc_model = (
    baseline[baseline["metric"]=="accuracy"]
    .pivot_table(
        index="run_name",
        columns="model",
        values="value"
    )
)

clean_acc_model_row = clean_acc_model.mean(axis=0)

print("üßº Clean Accuracy by model")
display(clean_acc_model_row)

clean_f1_model = (
    baseline[baseline["metric"]=="f1"]
    .pivot_table(
        index="run_name",
        columns="model",
        values="value"
    )
)

clean_f1_model_row = clean_f1_model.mean(axis=0)

print("üßº Clean F1 by model")
display(clean_f1_model_row)

clean_recall_model = (
    baseline[baseline["metric"]=="recall"]
    .pivot_table(
        index="run_name",
        columns="model",
        values="value"
    )
)

clean_recall_model_row = clean_recall_model.mean(axis=0)

print("üßº Clean Recall by model")
display(clean_recall_model_row)

```

```python
weak = df[
    (df["severity"] == "weak") &
    (df["modality"] != "__overall__")
].copy()

print("weak rows:", len(weak))
print("models:", weak["model"].unique())
print("run_name:", weak["run_name"].unique())
print("modalities:", weak["modality"].unique())

prompt_modality_summary = (
    weak
    .groupby(["run_name", "modality"], dropna=False)
    .agg(
        n_total=("n_total", "sum"),
        n_eval=("n_eval", "sum"),
        
        # ÏÑ±Îä•ÏßÄÌëú
        accuracy=("accuracy", "mean"),
        precision=("precision", "mean"),
        recall=("recall", "mean"),
        f1=("f1", "mean"),

        # üî• clean-set ÏÑ±Îä•
        baseline_accuracy=("baseline_accuracy", "mean"),
        
        # accuracy Í∏∞Î∞ò ÏÉÅÎåÄÏÑ±Îä•(%)
        performance_pct_change=("performance_pct_change", "mean")
    )
    .sort_values(["modality", "run_name"])
)

#display(prompt_modality_summary)
```

### Modality specific

```python
pivot_acc = (
    prompt_modality_summary
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="modality",
        values="accuracy"
    )
)

pivot_f1 = (
    prompt_modality_summary
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="modality",
        values="f1"
    )
)

pivot_recall = (
    prompt_modality_summary
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="modality",
        values="recall"
    )
)

pivot_base = (
    prompt_modality_summary
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="modality",
        values="baseline_accuracy"
    )
)

# modalityÎ≥Ñ clean baseline ÌèâÍ∑†
# clean_row = pivot_base.mean(axis=0)

# print("üßº Clean baseline row (per modality)")
# display(clean_row)

pivot_acc_with_clean = pd.concat(
    [
        pd.DataFrame([clean_acc_row], index=["clean_image"]),
        pivot_acc
    ],
    axis=0
)

print("üìä Weak Accuracy + Clean Baseline (from baseline_performance.csv)")
display(pivot_acc_with_clean)


pivot_f1_with_clean = pd.concat(
    [
        pd.DataFrame([clean_f1_row], index=["clean_image"]),
        pivot_f1
    ],
    axis=0
)

print("üìä Weak F1 + Clean Baseline (from baseline_performance.csv)")
display(pivot_f1_with_clean)


pivot_recall_with_clean = pd.concat(
    [
        pd.DataFrame([clean_recall_row], index=["clean_image"]),
        pivot_recall
    ],
    axis=0
)

print("üìä Weak Recall + Clean Baseline (from baseline_performance.csv)")
display(pivot_recall_with_clean)
```

```python
plot_heatmap(
    pivot_acc_with_clean,
    "Weak Accuracy by Prompt √ó Modality (incl. clean_image)"
)
plot_heatmap(
    pivot_f1_with_clean,
    "Weak F1 by Prompt √ó Modality (incl. clean_image)"
)
plot_heatmap(
    pivot_recall_with_clean,
    "Weak Recall by Prompt √ó Modality (incl. clean_image)"
)

```

### Model Specific

```python
pivot_acc_model = (
    weak
    .groupby(["run_name", "model"])
    .agg(accuracy=("accuracy", "mean"))
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="model",
        values="accuracy"
    )
)

pivot_f1_model = (
    weak
    .groupby(["run_name", "model"])
    .agg(f1=("f1", "mean"))
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="model",
        values="f1"
    )
)

pivot_recall_model = (
    weak
    .groupby(["run_name", "model"])
    .agg(recall=("recall", "mean"))
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="model",
        values="recall"
    )
)

pivot_acc_model_with_clean = pd.concat(
    [
        pd.DataFrame([clean_acc_model_row], index=["clean_image"]),
        pivot_acc_model
    ],
    axis=0
)

print("üìä Weak Accuracy + Clean Baseline (by model)")
display(pivot_acc_model_with_clean)

pivot_f1_model_with_clean = pd.concat(
    [
        pd.DataFrame([clean_f1_model_row], index=["clean_image"]),
        pivot_f1_model
    ],
    axis=0
)

print("üìä Weak F1 + Clean Baseline (by model)")
display(pivot_f1_model_with_clean)


pivot_recall_model_with_clean = pd.concat(
    [
        pd.DataFrame([clean_recall_model_row], index=["clean_image"]),
        pivot_recall_model
    ],
    axis=0
)

print("üìä Weak Recall + Clean Baseline (by model)")
display(pivot_recall_model_with_clean)
```

```python

plot_heatmap(
    pivot_acc_model_with_clean,
    "Weak Accuracy by Prompt √ó Model (incl. clean_image)"
)

plot_heatmap(
    pivot_f1_model_with_clean,
    "Weak F1 by Prompt √ó Model (incl. clean_image)"
)

plot_heatmap(
    pivot_recall_model_with_clean,
    "Weak Recall by Prompt √ó Model (incl. clean_image)"
)

```

# Strong

```python
strong = df[
    (df["severity"].astype(str).str.lower() == "strong") &
    (df["modality"].astype(str).str.lower() != "__overall__")
].copy()

print("strong rows:", len(strong))
print("run_name:", strong["run_name"].unique())
print("modalities:", strong["modality"].unique())
print("models:", strong["model"].unique())
```

```python
strong_prompt_modality = (
    strong
    .groupby(["run_name", "modality"], dropna=False)
    .agg(
        n_total=("n_total", "sum"),
        distorted_rate=("distorted_rate", "mean"),
    )
    .sort_values(["modality", "run_name"])
)


pivot_dist = (
    strong_prompt_modality
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="modality",
        values="distorted_rate"
    )
)

print("üìä Strong distorted_rate by Prompt √ó Modality")
display(pivot_dist)

plot_heatmap(
    pivot_dist,
    "Strong distorted_rate by Prompt √ó Modality"
)
```

```python
strong_prompt_model = (
    strong
    .groupby(["run_name", "model"], dropna=False)
    .agg(
        n_total=("n_total", "sum"),
        distorted_rate=("distorted_rate", "mean"),
    )
    .sort_values(["model", "run_name"])
)

pivot_dist_model = (
    strong_prompt_model
    .reset_index()
    .pivot_table(
        index="run_name",
        columns="model",
        values="distorted_rate"
    )
)

print("üìä Strong distorted_rate by Prompt √ó Model")
display(pivot_dist_model)

plot_heatmap(
    pivot_dist_model,
    "Strong distorted_rate by Prompt √ó Model"
)

```
