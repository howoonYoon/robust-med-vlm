```python
import pandas as pd

files = {
    "baseline": "qwen3vl_baseline_results.csv",
    "simple_cot": "qwen3vl_simple_cot_results.csv", #128
    "structured_cot": "qwen3vl_structured_cot_results.csv", #256
    "artifact_aware": "qwen3vl_artifact_aware_simple_results_.csv" #128
}

# 파일 경로
path_baseline = "qwen3vl_baseline_results.csv"
path_cot = "qwen3vl_simple_cot_results.csv"
path_scot =  "qwen3vl_structured_cot_results.csv"
path_art = "qwen3vl_artifact_aware_simple_results_.csv"

# CSV 로드
df_base = pd.read_csv(path_baseline)
df_cot  = pd.read_csv(path_cot)
df_scot = pd.read_csv(path_scot)
df_art  = pd.read_csv(path_art)

# simple_cot 파일에서 binarylabel 컬럼을 baseline에 그대로 순서대로 붙임
df_base["binarylabel"] = df_cot["binarylabel"].values

def normalize(df, model_name):
    df = df.copy()
    df["model_type"] = model_name
    df["model_label"] = df["model_label"].astype(str).str.strip().str.lower()

    label_map = {"normal": 0, "disease": 1}
    df["model_binary"] = df["model_label"].map(label_map)
    return df

df_base_n = normalize(df_base, "baseline")
df_cot_n  = normalize(df_cot, "simple_cot")
df_scot_n = normalize(df_scot, "structured_cot")
df_art_n  = normalize(df_art, "artifact_aware")
```

```python
df_all = pd.concat([df_base_n, df_cot_n,df_scot_n, df_art_n], ignore_index=True)
df_all["correct"] = (df_all["binarylabel"] == df_all["model_binary"])

acc_by_model = (
    df_all.groupby("model_type")["correct"]
          .mean()
          .reset_index()
          .sort_values("correct", ascending=False)
)

acc_by_model
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix
)
import numpy as np
import pandas as pd

def compute_metrics(df):
    y_true = df["binarylabel"]
    y_pred = df["model_binary"]

    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)
    
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp + 1e-9)

    return {
        "accuracy": acc,
        "precision": prec,
        "recall (sensitivity)": rec,
        "specificity": specificity,
        "f1_score": f1,
        "TP": tp,
        "TN": tn,
        "FP": fp,
        "FN": fn,
    }
```

```python
results = {}
for model_name in df_all["model_type"].unique():
    df_sub = df_all[df_all["model_type"] == model_name]
    results[model_name] = compute_metrics(df_sub)

pd.DataFrame(results).T
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np

def compute_metrics(df):
    y_true = df["binarylabel"]
    y_pred = df["model_binary"]

    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)

    # confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp + 1e-9)

    return {
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "specificity": specificity,
        "f1_score": f1,
        "TP": tp,
        "TN": tn,
        "FP": fp,
        "FN": fn
    }
```

```python
import matplotlib.pyplot as plt
from PIL import Image

def show_samples_by_dataset_severity(
    df, 
    dataset_name, 
    severity_value, 
    n_samples=4
):
    # 조건 필터링
    sub = df[(df["dataset"] == dataset_name) & (df["severity"] == severity_value)]

    if sub.empty:
        print(f"[{dataset_name} | severity={severity_value}] 데이터 없음")
        return
    
    # IDX 기준으로 샘플 선택
    idx_list = (
        sub["IDX"]
        .drop_duplicates()
        .sample(min(n_samples, sub["IDX"].nunique()), random_state=0)
        .tolist()
    )

    for idx in idx_list:
        df_sample = sub[sub["IDX"] == idx]

        gt = df_sample["binarylabel"].iloc[0]
        gt_text = "disease (1)" if gt == 1 else "normal (0)"
        fp = df_sample["original_filepath"].iloc[0]

        print("=" * 100)
        print(f"Dataset: {dataset_name} | severity={severity_value} | IDX={idx}")
        print("Image path:", fp)
        print("GT:", gt_text, "\n")

        # 모델별 예측 비교 테이블
        display(
            df_sample[[
                "model_type",
                "model_label",
                "model_binary",
                "correct",
                "model_reason"
            ]]
        )

        # 이미지 표시
        try:
            img = Image.open(fp).convert("RGB")
            plt.figure(figsize=(4, 4))
            plt.imshow(img)
            plt.axis("off")
            plt.title(f"{dataset_name} | sev={severity_value} | IDX={idx}\nGT: {gt_text}")
            plt.show()
        except Exception as e:
            print("❌ 이미지 로드 실패:", e)

import matplotlib.pyplot as plt
from PIL import Image

def show_samples_by_corruption_detail_sev2(
    df,
    n_samples=3
):
    # severity = 2 고정
    sub = df[df["severity"] == 2].copy()

    if sub.empty:
        print("severity=2 데이터 없음")
        return

    # corruption_detail 목록 추출
    corr_details = sorted(sub["corruption_detail"].dropna().unique())

    for corr in corr_details:
        sub_corr = sub[sub["corruption_detail"] == corr]

        if sub_corr.empty:
            continue

        print(f"\n######## corruption_detail = {corr} | severity = 2 ########\n")

        # IDX 기준으로 샘플 몇 개 선택
        idx_list = (
            sub_corr["IDX"]
            .drop_duplicates()
            .sample(min(n_samples, sub_corr["IDX"].nunique()), random_state=0)
            .tolist()
        )

        for idx in idx_list:
            df_sample = sub_corr[sub_corr["IDX"] == idx]

            gt = df_sample["binarylabel"].iloc[0]
            gt_text = "disease (1)" if gt == 1 else "normal (0)"
            fp = df_sample["original_filepath"].iloc[0]

            print("=" * 100)
            print(f"corruption_detail: {corr} | severity=2 | IDX={idx}")
            print("Image path:", fp)
            print("GT:", gt_text, "\n")

            # 모델별 세부 정보 출력
            display(
                df_sample[[
                    "model_type",
                    "model_label",
                    "model_binary",
                    "correct",
                    "model_reason"
                ]]
            )

            # 이미지 시각화
            try:
                img = Image.open(fp).convert("RGB")
                plt.figure(figsize=(4, 4))
                plt.imshow(img)
                plt.axis("off")
                plt.title(f"{corr} | sev=2 | IDX={idx}\nGT: {gt_text}")
                plt.show()
            except Exception as e:
                print("❌ 이미지 로드 실패:", e)

```
